5.In a single terror incident in India, find out top 5 cities by number killed and wounded

```python
terror['killnwound'] = terror['nkill']+terror['nwound']
terror.query('country_txt=="India"').groupby('city').agg({'killnwound':'sum'}).sort_values(by='killnwound',ascending=False).head(6)
```

***top 5 cities by number killed and wounded in a single terror incident in India***

7.label all the incidents where the number killed was more than 5 as severe. 

```python
severe=[]
for i in terror['nkill']:
    if i>=5:
        severe.append('severe')
    else:
        severe.append('nonsever')
terror['severe'] = severe
terror.head()
```

***above output shows all the incidents and number killed more than 5 labelled as "severe" in the column severe***

8.write a function to label an incident that was both successful and suicidal

```python
def labelincident(terror,eventid):
    dict1 =terror[terror['eventid']==eventid]
    print(dict1['success'],':',dict1['suicide'])
    if (int(dict1['success'])==1) & (int(dict1['suicide'])==1):
        return True
    else:
        return False
labelincident(terror,int(input("Enter an event id")))
```

***above function returns the incidents which were both successful and suicidal by giving the eventid***

9.Create a new category representing if the incident occured in Afghanistan, Pakistan or India as one level of the category and all the other countries as another level

```python
countries = list(terror['country_txt'])
category = []
countrytrue = ['India','Pakistan','Afghanistan']
for country in countries:
    if country in countrytrue:
        category.append('1')
    else:
        category.append('0')
terror['category'] = category
terror.head()
```

***above represent a new category of countries
1 level:if the incident occured in Afghanistan, Pakistan or India 
0 level: for all the other countries***

11.List the number of suicides attacks and average kills by Af-Pak-India vs ROW. Rename columns in the output as Average_Kills and Number_Incidents.

```python
terror.groupby('category').agg({'suicide':'count','nkill':'mean'}).rename(columns={'suicide':'Number_Incident','nkill':'Average_Kills'})
```
