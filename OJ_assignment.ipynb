12.Add a new column to the dataframe normInc which is the normalised value of income

```python
oj_data['normInc'] = ((oj_data['income'])-oj_data['income'].mean())/(oj_data['income'].max()-oj_data['income'].min())
```

***inorder to find the normalised income the formula used is
Normalised income= (income-mean(income))/(max(income)-min(income))
the caluculated normalised income is stored as new column in the dataframe with the name normInc***
